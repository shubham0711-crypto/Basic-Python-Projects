```python
data = data.split('\n\n\n')
new_data = []
for ques in data :
    print(repr(ques),end='\n\n\n\n')
```

```python
c = 1
q_list = []
o_list = []
a_list = []
for dt in data : 
    d = dt
    d = d.split('\n')
    ques = d[0]
    q_list.append(ques)
    option = d[1:5]
    #option = list(zip(['A.','B.','C.','D'],option))
    o_list.append(option)
    ans = d[-1]
    a_list.append(ans)
```

```python
c = 1
for q in q_list : 
    print(c,q.replace('\n',''),'\n\n')
    c = c + 1
    
```

```python
c = 1
for options in o_list:
    print("Options for Question : ",c)
    op = [ 'A. ','B. ','C. ','D. ']
    op_data = zip(op,options)
    for ch,ops in op_data :
        print(ch,ops,sep='',end='\t')
    print('\n\n')
    c = c + 1
```

```python
c = 1
for ans in a_list:
    print("Answer for Question : ",c)
    print(ans,end='\n\n')
    c = c + 1
```

```python
print("Total Question : ",len(q_list))
q_bank =  { }
c = 1
print("Adding Questions")
for var in range(len(q_list)) :
    print(c,end=' ')
    if q_list[var] in q_bank :
        print("Repeated Questions : ")
        print(q_list[var])
    q_bank[q_list[var]] = (o_list[var],a_list[var])
    c = c + 1
```

```python
from pprint import pprint
pprint(q_bank)
print("\nTotal Questions ",len(q_bank))
```

```python
c = 1
for ques in q_bank:
    print(c,ques)
    print()
    l = [ '1. ','2. ','3. ','4. ']
    for v in range(4):
        print(l[v],q_bank[ques][0][v],sep='',end='\t')
    print('\n')
    print(q_bank[ques][1])
    print('\n\n')

    c  = c + 1
```

```python
print("Writing Data to Question Bank")
import json
f = open('question_bank.db','w')
json.dump(q_bank,f)
f.close()
print("Question Bank is Ready to use Now")
```
